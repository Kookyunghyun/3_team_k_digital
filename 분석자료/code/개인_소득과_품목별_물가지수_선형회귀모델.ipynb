```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

for i in lst:
  print(i)
  df2.insert(1,i,list(df1[i]))

  data_price_index=df2.loc[:,i]
  data_income = df2.loc[:,'1인당 개인소득']

  income= np.array(data_income)
  price_index = np.array(data_price_index)

  #훈련 세트와 테스트 세트로 나눔
  train_input, test_input, train_target, test_target = train_test_split(income, price_index, random_state=42)

  #훈련 세트와 테스트 세트를 2차원 배열로 바꿈
  train_input = train_input.reshape(-1,1)
  test_input = test_input.reshape(-1,1)

  lr = LinearRegression()
  lr.fit(train_input,train_target)

  # 훈련 세트의 산점도
  plt.scatter(train_input, train_target)

  # 10000~22000 값사이의 1차 방정식 그래프
  plt.plot([10000,22000],[10000*lr.coef_+lr.intercept_, 22000*lr.coef_+lr.intercept_])

  sorted(train_target)
  new_train_input = train_input.astype(int)
  n_train_input = sorted(new_train_input)
  lste = []
  for y in range(0,15):
    lste.append(n_train_input[y][0])
  sorted(lste)
  y_values = []

  for x in lste:
    y = 0.00255131 * x + 52.26802890806545
    y_values.append(y)

  a = sorted(train_target)
  s = sorted(y_values)

  MSE = np.square(np.subtract(a,s)).mean()

  # 회귀식 출력
  title_graph= (f"회귀식:y = {lr.coef_}x + {lr.intercept_}")
  plt.title(title_graph + "\n" + "MSE:" + str(MSE))
  plt.xlabel('1인당 개인소득')
  plt.ylabel('물가지수')
  plt.savefig(i)
  plt.clf()

  df2 = df2.drop(i,axis=1)
```
