```python
lst = list(df.index)
```

```python
def find_duplicates(lst):
    counts = {}
    duplicates = []
    for item in lst:
        if item in counts:
            if counts[item] == 1:
                duplicates.append(item)
            counts[item] += 1
        else:
            counts[item] = 1
    return duplicates

duplicates = find_duplicates(lst)
print(duplicates)
```

```python
for i in duplicates:
    df = df.drop(i,axis=0)
```

```python
df = df.drop("품목별",axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D


encode = "EUC-KR"
# 한글 폰트 경로 설정
font_path = 'C:/Windows/Fonts/NGULIM.TTF'

# 폰트 이름 얻어오기
font_name = fm.FontProperties(fname=font_path).get_name()
# 컬럼명 추출
columns = df1.columns
plt.rcParams['font.family'] = font_name

df1.index = pd.to_datetime(df1.index) # 날짜를 인덱스로 설정
# SARIMA 모델 생성 및 학습
for col in columns:
    model = sm.tsa.SARIMAX(df1[col], order=(1, 0, 0), seasonal_order=(1, 0, 0, 12))
    results = model.fit()

    # 예측
    forecasted_values = results.get_forecast(steps=3)  # 다음 3개 값 예측
    forecasted_data = forecasted_values.summary_frame()  # 예측 결과를 데이터프레임으로 변환

    # 그래프 표시
    plt.figure(figsize=(10, 5))
    plt.plot(df1.index, df1[col], label='실제값')
    plt.plot(forecasted_data.index, forecasted_data['mean'], label='예측값')
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.title(f'{col} Forecast')
    plt.legend()
    plt.show()
```
