```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

sequence_length = 10  # 입력 시퀀스 길이
X = []
y = []
for i in range(len(df) - sequence_length):
    X.append(df.iloc[i:i+sequence_length]['개인소득'].values)
    y.append(df.iloc[i+sequence_length]['물가지수'])
X = np.array(X)
y = np.array(y)


scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))

```

```python
# 모델 구성
model = Sequential()
model.add(LSTM(64, input_shape=(sequence_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# 모델 훈련
model.fit(X_scaled, y_scaled, epochs=100, batch_size=128)

# 개인소득이 25000일 때의 물가지수 예측
def pred(num):
    income = np.array([num])  # 개인소득이 25000인 경우
    income_scaled = scaler.transform(income.reshape(-1, 1))
    X_test = np.append(X_scaled[-1][1:], income_scaled)  # 마지막 4개의 개인소득과 25000을 합침
    X_test = X_test.reshape(1, sequence_length, 1)
    predicted_scaled = model.predict(X_test)
    predicted = scaler.inverse_transform(predicted_scaled)
    return predicted

pred(25000)
```

```python
# 2022년 예측
last_5_years = df['물가지수'][-19:].values.reshape(1, -1, 1)
predicted_scaled = model.predict(last_5_years)
predicted = scaler.inverse_transform(predicted_scaled)
print('2022년 물가지수 예측:', predicted[0][0])
```

```python
# 초기 입력 시퀀스
initial_sequence = df['물가지수'][-19:].values.reshape(1, -1, 1)

# 향후 5년간의 예측
predictions = []
current_sequence = initial_sequence
for _ in range(5):
    predicted_scaled = model.predict(current_sequence)
    predicted = scaler.inverse_transform(predicted_scaled)
    predictions.append(predicted[0][0])
    current_sequence = np.append(current_sequence[:, 1:, :], predicted_scaled.reshape(1, 1, 1), axis=1)

# 예측 결과 출력
for i, prediction in enumerate(predictions):
    year = int(df.index[-1]) + i + 1
    print(f"{year}년 물가지수 예측: {prediction}")
```
